```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
age = np.array([74,74,74,74,74,74,38,38,38,16,16,16,16,16,16])
area=np.array([16.1,15.5,14.5,17.2,14.9,17.2,15.8,15.4,17.4,17.7,14.9,15.8,15.8,14.2,17.4])
ar=np.array([3.894,4.048,4.031,3.837,4.032,4.114,4.098,4.304,4.049,3.884,3.930,3.874,3.837,3.904,3.805])
sf=np.array([0.549,0.551,0.558,0.544,0.557,0.536,0.541,0.526,0.524,0.543,0.558,0.551,0.551,0.557,0.558])
dict = {'Age':age,
        'Area':area,
        'Aspect Ratio':ar,
        'Circularity':sf}
df = pd.DataFrame(dict)
```

```python
fig,ax = plt.subplots(figsize=(5,10))
ax0 = sns.boxplot(x="Age", y="Area", data=df)
ax0 = sns.swarmplot(x="Age", y="Area", data=df, edgecolor=(247/255, 69/255, 13/255), linewidth=4)
ax0.set_xlabel('Age',fontsize=20)
ax0.set_ylabel('Area',fontsize=20)
ax0.tick_params(axis='both',labelsize=20)
fig.savefig('temp.svg')
```

```python
fig,ax = plt.subplots(figsize=(5,10))
ax0 = sns.boxplot(x="Age", y="Aspect Ratio", data=df)
ax0 = sns.swarmplot(x="Age", y="Aspect Ratio", data=df, edgecolor=(247/255, 69/255, 13/255), linewidth=4)
ax0.set_xlabel('Age',fontsize=20)
ax0.set_ylabel('Aspect Ratio',fontsize=20)
ax0.tick_params(axis='both',labelsize=20)
fig.savefig('temp.svg')
```

```python
fig,ax = plt.subplots(figsize=(5,10))
ax0 = sns.boxplot(x="Age", y="Circularity", data=df)
ax0 = sns.swarmplot(x="Age", y="Circularity", data=df, edgecolor=(247/255, 69/255, 13/255), linewidth=4)
ax0.set_xlabel('Age',fontsize=20)
ax0.set_yticks([0.52,0.53,0.54,0.55,0.56])
ax0.set_ylabel('Circularity',fontsize=20)
ax0.tick_params(axis='both',labelsize=20)
fig.savefig('temp.svg')
```
