```python
from openslide import OpenSlide
import os
```

```python
svs_src = r'\\kukissd\Kyu_Sync\Aging\data\svs'
svss = [os.path.join(svs_src,_) for _ in os.listdir(svs_src) if _.endswith('svs')]
downsample_factor = 16.1
svs_dst = os.path.join(svs_src,'level_2')
if not os.path.exists(svs_dst): os.mkdir(svs_dst)
```

```python
for svs in svss:
    svs_obj = OpenSlide(svs)
    level = svs_obj.get_best_level_for_downsample(downsample_factor)
    print('level for downsample factor {:.1f} is {:d}'.format(downsample_factor,level))
    svs_img = svs_obj.read_region(location=(0,0),level=level,size=svs_obj.level_dimensions[level]).convert('RGB')
    src,fn = os.path.split(svs)
    fn,ext = os.path.splitext(fn)
    fn = fn + '.tif'
    svs_img.save(os.path.join(svs_dst,fn))
```
